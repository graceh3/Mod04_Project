### 0. Import all necessary libraries

```python
import numpy as np
import time
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import urllib
import pandas as pd
import cv2
```

### 1. Get image link for all products

```python
urls = ['https://www.livefashionable.com/collections/totes',
       'https://www.livefashionable.com/collections/backpacks',
       'https://www.livefashionable.com/collections/crossbody-bags',
       'https://www.livefashionable.com/collections/satchels-bucket-bags']
```

```python
page = requests.get('https://www.livefashionable.com/collections/totes')
soup = BeautifulSoup(page.content, 'lxml')

match = soup.findAll('span', {'class':'grid-link__image-centered reveal'}) # find all tags with images
link = ['http:'+ x.img['src'] for x in match]
match_ = soup.findAll('div', {'class': 'product_info'})
des =[x.p.text for x in match_]
match_2 = soup.findAll('span', {'class': 'money'})
price = [x.text for x in match_2]
```

```python
link
```

```python
des
```

```python
links =[]
desc=[]
prices=[]

for url in urls:
    
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'lxml')

    match = soup.findAll('span', {'class':'grid-link__image-centered reveal'}) # find all tags with images
    links.extend(['http:'+ x.img['src'] for x in match])
    match2 = soup.findAll('div', {'class': 'product_info'})
    desc.extend([x.p.text for x in match2])

    match3 = soup.findAll('span', {'class': 'money'})
    prices.extend([x.text for x in match3])
```

```python
print(len(links))
print(len(desc))
print(len(prices))
```

```python
prices=[p for p in prices if p!= '$ 0.00 ']
```

```python
len(prices)
```

```python
# create dictionary with all item details to put into dataframe
all_data = []

for i in range(len(links)):
    all_data.append({'description': desc[i], 
                     'link': links[i], 
                     'price': prices[i], 
                    'brand': 'Able'})
```

```python
# list of dictionary in dataframe
df_data = pd.DataFrame(all_data)
df_data
```

```python
df_data.drop_duplicates(subset='description', inplace = True)
df_data
```

```python
df_data.reset_index(inplace = True)
df_data.drop(labels='index', axis = 1, inplace = True)
df_data
```

```python
num = list(range(len(df_data)))
num
```

```python
df_data = pd.concat([df_data, pd.Series(num)], axis = 1)
```

```python
df_data['img-file'] = df_data[0].apply(lambda x: 'able_'+str(x+1)+'.png')
```

```python
df_data
```

```python
df_data.drop(labels = [0], axis = 1, inplace = True)
df_data
```

```python
# save the dataframe as a csv (to merge with other data later)
df_data.to_csv('able_data.csv')
```

### 2. Loop through image links and save as a .png file

```python
# loop through links to request and save image locally
for i, url in enumerate(df_data.link):
    time.sleep(1)
    try:
        request = urllib.request.Request(url)
        response = urllib.request.urlopen(request)
        binary_str = response.read()
        byte_array = bytearray(binary_str)
        numpy_array = np.asarray(byte_array, dtype='int8')
        image = cv2.imdecode(numpy_array, cv2.IMREAD_UNCHANGED)
        cv2.imwrite("AmourVert/"+"able_{}".format(i+1)+".png", image)
        print("Saved "+"able_{}".format(i+1)+".png")
    except Exception as e:
        print(str(e))

print("Done")
```
