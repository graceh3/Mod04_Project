### 0. Import all necessary libraries

```python
import numpy as np
import time
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import urllib
import pandas as pd
import cv2
```

### 1. Get image link for all products

```python
urls = ["https://graine.com.au/collections/handbags?page=1",
       "https://graine.com.au/collections/handbags?page=2",
       "https://graine.com.au/collections/side-bags"]
```

```python
links = []
desc = []
prices = []

for url in urls:
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'lxml')
    match = soup.findAll('a', {'class':'grid-view-item__link grid-view-item__image-container'}) # find all tags with images
    links.extend(['http:'+ x.div.div.img['src'] for x in match])
    match2 = soup.findAll('div', {'class':'h4 grid-view-item__title'})
    desc.extend([x.text for x in match2])
    match3 = soup.findAll('span', {'class': 'product-price__price'})
    prices.extend([x.text for x in match3])
```

```python
print(len(prices))
print(len(desc))
print(len(links))
```

```python
prices
```

```python
pattern = '\d+.\d+'
p = re.compile(pattern)
prices = [p.findall(x) for x in prices]
prices = [float(x[0]) for x in prices]
prices
```

```python
# create dictionary with all item details to put into dataframe
all_data = []

for i in range(len(links)):
    all_data.append({'description': desc[i], 
                     'link': links[i], 
                     'price': prices[i], 
                    'brand': 'Graine',
                    'num': i})
```

```python
# list of dictionary in dataframe
df_data = pd.DataFrame(all_data)
df_data
```

```python
df_data['img-file'] = df_data['num'].apply(lambda x: 'graine_'+str(int(x)+1)+'.png')
```

```python
df_data
```

```python
df_data.drop(columns = ['num'], axis = 1, inplace = True)
df_data
```

```python
# save the dataframe as a csv (to merge with other data later)
df_data.to_csv('graine_data.csv')
```

### 2. Loop through image links and save as a .png file

```python
# loop through links to request and save image locally
for i, url in enumerate(df_data.link):
    time.sleep(1)
    try:
        request = urllib.request.Request(url)
        response = urllib.request.urlopen(request)
        binary_str = response.read()
        byte_array = bytearray(binary_str)
        numpy_array = np.asarray(byte_array, dtype='int8')
        image = cv2.imdecode(numpy_array, cv2.IMREAD_UNCHANGED)
        cv2.imwrite("Graine/"+"graine_{}".format(i+1)+".png", image)
        print("Saved "+"graine_{}".format(i+1)+".png")
    except Exception as e:
        print(str(e))

print("Done")
```
