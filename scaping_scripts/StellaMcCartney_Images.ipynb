### 1. Get image link for all products

```python
url = "https://www.stellamccartney.com/us/online/stella-mccartney/women/handbags#{%22ytosQuery%22:%22true%22,%22department%22:%22main_bags%22,%22gender%22:%22D%22,%22season%22:%22A,P,E%22,%22yurirulename%22:%22searchwithdepartment%22,%22page%22:%223%22,%22productsPerPage%22:%2240%22,%22suggestion%22:%22false%22,%22facetsvalue%22:[],%22totalPages%22:3,%22rsiUsed%22:%22false%22,%22totalItems%22:%22100%22,%22partialLoadedItems%22:%2240%22,%22itemsToLoadOnNextPage%22:%2240%22}"
```

```python
browser = webdriver.Chrome(ChromeDriverManager().install())

# Tell Selenium to get the URL you're interested in.
browser.get(url)
time.sleep(5)

# Selenium script to grab all product info on page -> wait 3 seconds for the next batch of data to load, 
# then continue scrolling. It will continue to do this until the page stops loading new data.
all_ = []
links_ = []
desc_ = []
prices_ = []


elements_ = (browser.find_elements_by_class_name(name= 'image'))
links_.extend([e.get_attribute('innerHTML') for e in elements_])
elements = (browser.find_elements_by_class_name(name= 'mainImage'))
desc_.extend([e.get_attribute('alt') for e in elements])
elements2 = (browser.find_elements_by_class_name(name= 'value'))
prices_.extend([e.text for e in elements2])

    
# Now that the page is fully scrolled, grab the source code.
source_data = browser.page_source
browser.close()
```

```python
print(len(links_))
print(len(desc_))
print(len(prices_))
```

```python
links_
```

```python
links_f = []
for l in links_:
    groups = re.search('.*?src\=\"(.*)c\.jpg\"', l)
    links_f.append((groups.group(1)+'c.jpg').replace('https','http'))
```

```python
links_f
```

```python
prices_ = prices_[1:101]
```

```python
print(len(prices_))
print(len(links_f))
print(len(desc_))
```

```python
# create dictionary with all item details to put into dataframe
all_data = []

for i in range(len(links_f)):
    all_data.append({'description': desc_[i], 
                     'link': links_f[i], 
                     'price': float(prices_[i].replace(",","")), 
                    'brand': 'Stella McCartney',
                    'num': i})
```

```python
# list of dictionary in dataframe
df_data = pd.DataFrame(all_data)
df_data
```

```python
df_data['img-file'] = df_data['num'].apply(lambda x: 'stella_'+str(int(x)+1)+'.jpg')
```

```python
df_data
```

```python
# save the dataframe as a csv (to merge with other data later)
df_data.to_csv('stella_data.csv')
```

### 2. Loop through image links and save as a .png file

```python
# loop through links to request and save image locally
from PIL import Image
import requests
from io import BytesIO

for i, url in enumerate(df_data.link):
    try:
        response = requests.get(url)
        img = Image.open(BytesIO(response.content))
        image_save_loc = 'Stella/'+'stella_'+str(int(i)+1)+'.jpg'
        img.save(image_save_loc, "JPEG")
        print("Saved "+'stella_'+str(int(i)+1)+".jpg")
    except Exception as e:
        print(str(e))

print("Done")
```
