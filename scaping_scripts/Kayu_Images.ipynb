# Scraping Kayu product images
***

### 0. Import all necessary libraries

```python
import numpy as np
import time
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import urllib
import pandas as pd
import cv2
```

### 1. Get image link for all products

```python
url = "https://shop.kayudesign.com/collections/all/tote"
```

```python
links = []
desc = []
prices = []

page = requests.get(url)
soup = BeautifulSoup(page.content, 'lxml')
match = soup.findAll('div', {'class':'product-index'}) # find all tags with images
links = ['http:'+ x.img['src'] for x in match]
desc=[x.img['alt'] for x in match]
match2 = soup.findAll('div', {'class': 'price'})
prices = [x.text for x in match2]
```

```python
len(prices)
```

```python
prices
```

```python
pattern = '\d+.\d+'
p = re.compile(pattern)
prices = [p.findall(x) for x in prices]
prices = [float(x[0]) for x in prices]
prices
```

```python
# create dictionary with all item details to put into dataframe
all_data = []

for i in range(len(links)):
    all_data.append({'description': desc[i], 
                     'link': links[i], 
                     'price': prices[i], 
                    'brand': 'Kayu',
                    'num': i})
```

```python
# list of dictionary in dataframe
df_data = pd.DataFrame(all_data)
df_data
```

```python
df_data['img-file'] = df_data['num'].apply(lambda x: 'kayu_'+str(int(x)+1)+'.png')
```

```python
df_data
```

```python
df_data.drop(columns = ['num'], axis = 1, inplace = True)
df_data
```

```python
# save the dataframe as a csv (to merge with other data later)
df_data.to_csv('kayu_data.csv')
```

### 2. Loop through image links and save as a .png file

```python
# loop through links to request and save image locally
for i, url in enumerate(df_data.link):
    time.sleep(1)
    try:
        request = urllib.request.Request(url)
        response = urllib.request.urlopen(request)
        binary_str = response.read()
        byte_array = bytearray(binary_str)
        numpy_array = np.asarray(byte_array, dtype='int8')
        image = cv2.imdecode(numpy_array, cv2.IMREAD_UNCHANGED)
        cv2.imwrite("Kayu/"+"kayu_{}".format(i+1)+".png", image)
        print("Saved "+"kayu_{}".format(i+1)+".png")
    except Exception as e:
        print(str(e))

print("Done")
```
