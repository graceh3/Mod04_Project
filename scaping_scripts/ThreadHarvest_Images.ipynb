### 1. Get  name and price for all products w/ BeautifulSoup

```python
urls = ["https://threadharvest.com.au/collections/bags-1?page=1",
       "https://threadharvest.com.au/collections/bags-1?page=2",
       "https://threadharvest.com.au/collections/bags-1?page=3"]

desc = []
prices = []
```

```python
for url in urls:
    
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'lxml')

    match2 = soup.findAll('h2', {'class':'ProductItem__Title'})
    desc.extend([x.text for x in match2])

    match3 = soup.findAll('span', {'class':'ProductItem__Price'}) 
    prices.extend([x.text for x in match3])
```

### Selenium to get image links


```python
from selenium import webdriver
!pip install webdriver_manager
from webdriver_manager.chrome import ChromeDriverManager
```

```python
links = []
```

```python
for url in urls:
    
    browser = webdriver.Chrome(ChromeDriverManager().install())

    browser.get(url)
    time.sleep(5)

    # Selenium script to grab all product info on page -> wait 3 seconds for the next batch of data to load, 
    # then continue scrolling. It will continue to do this until the page stops loading new data.

    elements = (browser.find_elements_by_class_name(name= 'ProductItem__Image'))
    links.extend([e.get_attribute('data-srcset') for e in elements])


    # Now that the page is fully scrolled, grab the source code.
    source_data = browser.page_source
    browser.close()
```

### Combine all data into a dictionary, then dataframe

```python
# create dictionary with all item details to put into dataframe
all_data = []

for i in range(len(desc_)):
    all_data.append({'description': desc_[i], 
                     'price': prices[i], 
                    'brand': 'Thread Harvest',
                    'num': i})
```

```python
# list of dictionary in dataframe
df_data = pd.DataFrame(all_data)
df_data
```

```python
df_data['img-file'] = df_data['num'].apply(lambda x: 'th_'+str(int(x)+1)+'.jpg')
```

```python
df_data
```

```python
df_data.drop(columns = ['num'], axis = 1, inplace = True)
df_data
```

```python
df_data.price = df_data.price.apply(lambda x: x.replace('$', ""))
df_data.price = df_data.price.apply(lambda x: float(x.replace('From', "")))
```

```python
# save the dataframe as a csv (to merge with other data later)
df_data.to_csv('thread_data.csv')
```

### 2. Loop through image links and save as a .png file

```python
# loop through links to request and save image locally
from PIL import Image
import requests
from io import BytesIO

for i, url in enumerate(links):
    try:
        response = requests.get(url)
        img = Image.open(BytesIO(response.content))
        image_save_loc = 'ThreadHarvest/'+'ek_'+str(int(i)+1)+'.png'
        img.save(image_save_loc, "PNG")
        print("Saved "+'th_'+str(int(i)+1)+".png")
    except Exception as e:
        print(str(e))

print("Done")
```
