### 0. Import all necessary libraries

```python
import numpy as np
import time
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import urllib
import pandas as pd
import cv2
```

### 1. Get image link for all products

```python
url = "https://www.elvisandkresse.com/collections/womens-bags"
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.content, 'lxml')
soupy = soup.find('div', {'class': 'products'})

match = soupy.findAll('img', {'class':'lazyloaded'}) # find all tags with images
links = ['http:'+ x['src'] for x in match]

match2 = soupy.findAll('span', {'class':'title'}) 
desc = [x.text for x in match2]


    
```

```python
desc
```

```python
prices = [260.69,
         320.95,
         235.80,
         275.10,
         353.71,
         340.61,
         379.91,
         314.40,
         229.25,
         196.50,
         209.60,
         248.90,
         196.5,
         209.6,
         301.3,
         117.9,
         327.5,
         327.5,
         392.99]
```

```python
print(len(prices))
print(len(links))
```

```python
# create dictionary with all item details to put into dataframe
all_data = []

for i in range(len(links)):
    all_data.append({'description': desc[i], 
                     'link': links[i], 
                     'price': prices[i], 
                    'brand': 'Elvis & Kresse',
                    'num': i})
```

```python
# list of dictionary in dataframe
df_data = pd.DataFrame(all_data)
df_data
```

```python
df_data['img-file'] = df_data['num'].apply(lambda x: 'ek_'+str(int(x)+1)+'.jpg')
```

```python
df_data
```

```python
df_data.drop(columns = ['num'], axis = 1, inplace = True)
df_data
```

```python
# save the dataframe as a csv (to merge with other data later)
df_data.to_csv('ek_data.csv')
```

### 2. Loop through image links and save as a .png file

```python
# loop through links to request and save image locally
from PIL import Image
import requests
from io import BytesIO

for i, url in enumerate(df_data.link):
    try:
        response = requests.get(url)
        img = Image.open(BytesIO(response.content))
        image_save_loc = 'Elvis/'+'ek_'+str(int(i)+1)+'.jpg'
        img.save(image_save_loc, "JPEG")
        print("Saved "+'ek_'+str(int(i)+1)+".jpg")
    except Exception as e:
        print(str(e))

print("Done")
```
